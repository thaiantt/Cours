## EXERCICE 1

```python
# Redefine rho_alpha
# params : 
# x : array
# alpha : float

def rho_alpha(X, alpha):
    if alpha == 0:
        f = np.abs(X)
    else:
        M = np.abs(X) < alpha
        f = (alpha * np.abs(X) - (alpha ** 2) /2)
        f[M] =  ((X ** 2) /2)[M]
    return f
```

## Question 1

```python
abs_min = -5
abs_max = 5

alphas = [0, 0.5, 2, 5]

x_q1 = np.linspace(abs_min, abs_max)

y_q1 = [rho_alpha(x_q1, alpha) for alpha in alphas]

for i in range(len(y_q1)):
    plt.plot(x_q1, y_q1[i], linewidth=lw, alpha=alp,
             label=r"$\alpha$ = {}".format(alphas[i]))

plt.xlabel("x")
plt.ylabel(r"$\rho_\alpha (x)$")

plt.title(r"$\rho_\alpha$ for different $\alpha$ values : {}".format(alphas))
plt.legend()
plt.show()
```

Pour $\alpha$ grand on se rapproche d'une m√©thode des moindres carr√©s classique. Pour des valeurs de $\alpha$ faibles, c'est-√†-dire proches de z√©ro, on se rapproche d'une m√©thode de type LAD (Least Absolute Deviation).

## Question 2

```python
n = 100
d = 2 # size of x column

theta_star = np.array([1, 1, 1]).reshape(-1,1)
print(n, d, theta_star.shape)
```

```python
# Create n vectors
X = np.random.uniform(size=(n,d))

print(X.shape)
print(type(X))
```

```python
# epsilon_i
Eps = np.random.normal(0, 1, (n,1)) # params : center, standard deviation, size

print(Eps.shape)
print(type(Eps))
```

```python
# y_i
x_theta = np.dot(X, theta_star[1:])

Y = np.array(theta_star[0] + x_theta + Eps)

print(Y.shape)
print(type(Y))
```

## Question 3

```python
init = np.array([0, 0, 0]).reshape(-1,1)

def run_minimize(X, Y, alpha, theta_init=init):

    def f_to_minimize(theta, alpha):
        t0 = theta[0].reshape(-1,1)
        t = theta[1:].reshape(-1,1)
        rho_alpha_computed = rho_alpha(Y - t0 - np.dot(X, t), alpha)
        return np.sum(rho_alpha_computed)

    theta_hat_alpha = scipy.optimize.minimize(
        f_to_minimize, x0=theta_init, args=alpha, method='nelder-mead', options={'xtol': 1e-5})
    return theta_hat_alpha
```

```python
# Test with alpha = 2
alpha_test = 2
print("TEST WITH ALPHA = 2")
minimized = run_minimize(X, Y, alpha_test, theta_init=init)

print(minimized.message)
print("     | Termination status of the optimizer : {}".format(
    minimized.status))
print("     | Number of iterations : {}".format(minimized.nit))
print("     | Value of Objective function : {}".format(minimized.fun))
print()
print("Solution [ Theta0 Theta1[0] Theta1[1] ] : {}".format(
    minimized.x))

theta_hat = minimized.x
```

## Question 4

```python
# unit epsilon sample
def create_sample(X, Y, theta_hat):
    t0 = theta_hat[0].reshape(-1,1)
    t = theta_hat[1:].reshape(-1,1)
    Eps_hat = Y.reshape(-1,1) - t0 - np.dot(X, t)

    eps_sample = np.random.choice(
        Eps_hat.ravel(), size=(len(Eps_hat.ravel()), 1), replace=True)

    return eps_sample
```

```python
# Compute bootstrap estimator for alpha = 2
def compute_estimator(X, Y, alpha, theta_hat, theta_init=init):

    t0 = theta_hat[0].reshape(-1,1)
    t = theta_hat[1:].reshape(-1,1)
    eps_sample_b = create_sample(X, Y, theta_hat)
    
    y_star = t0 + np.dot(X, t) + eps_sample_b

    results_boot = run_minimize(X, y_star, alpha, theta_init=theta_init).x
    return results_boot


print("Bootstrap estimator Theta_alpha_star :",
      compute_estimator(X, Y, alpha_test, theta_hat))
```

## Question 5

```python
B = 200
```

```python
def generate_thetas(B, X=X, Y=Y, theta_hat=theta_hat, theta_init=init, alpha = 2):
    thetaBootstrapList = []
    for i in range(B):
        thetaBootstrapList.append(compute_estimator(X, Y, alpha, theta_hat, theta_init=theta_init))
    thetaBootstrapList = np.array(thetaBootstrapList)
    results = thetaBootstrapList
    return results
```

```python
results = generate_thetas(B)
print(type(results))
print(results.shape)
```

```python
Vboot = np.cov(results, rowvar=False)
print(Vboot)
```

```python
theta_hat = np.array([1, 1, 1])
bboot = np.sum((results - theta_hat), axis=0) / B

print(bboot)
```

```python
def compute_EQMboot_alpha(results, theta_hat):
    meanboot = np.sum(results, axis=0) / B
    term_to_norm = results - meanboot
    term_normed = np.linalg.norm(term_to_norm, ord=2, axis=1)**2
    norm_b = np.sum(term_normed) / B
    norm_alpha = np.linalg.norm(theta_hat - meanboot, ord=2)**2
    
    EQMboot_alpha = norm_alpha + norm_b  # Still need to add theta alphas
    return EQMboot_alpha

print(compute_EQMboot_alpha(results, theta_hat))
```

## Question 6

```python
# Alpha grid for plot
alpha_grid = np.linspace(0, 10, num=20)


def find_alpha_min(X, Y, theta_init=init, Eps=Eps, alpha_grid=alpha_grid, B=B):
    EQMs = []

    for alpha_fix in alpha_grid:
        print("Alpha = ", alpha_fix)
        theta_hat_alp = run_minimize(X, Y, alpha_fix, theta_init=theta_init).x
        theta_bootstrap = compute_estimator(
            X, Y, alpha_fix, theta_hat_alp, theta_init=theta_init)
        bootstrap_list = generate_thetas(
            B, X=X, Y=Y, theta_hat=theta_hat_alp, theta_init=theta_init, alpha=alpha_fix)
        results = np.array(bootstrap_list)
        EQM = compute_EQMboot_alpha(results, theta_hat_alp)
        print("     | EQM = ", EQM)
        EQMs.append(EQM)
    return EQMs
```

```python
EQMs = find_alpha_min(X, Y, Eps=Eps, alpha_grid=alpha_grid, B=B)
```

```python
fig, ax = plt.subplots()

ax.plot(alpha_grid, EQMs, c='mediumpurple', marker='o',
           linewidth=lw, alpha=alp, label=r"$EQM_{boot}(\alpha)$")

min_index = np.argmin(EQMs)

ax.scatter(alpha_grid[min_index], EQMs[min_index],
           marker="x", s=300, linewidth=3, label="Minimum")
ax.axvline(alpha_grid[min_index], linestyle="--", c='grey', alpha=alp)
ax.axhline(EQMs[min_index], linestyle="--", c='grey', alpha=alp)
ax.text(alpha_grid[min_index], EQMs[min_index] , s="( {}, {})".format(round(alpha_grid[min_index],
                                                                                  2), round(EQMs[min_index], 2)), verticalalignment="top", horizontalalignment="center", fontsize=12)

plt.title(r"$EQM_{boot}(\alpha)$ for different $\alpha$ values")
plt.xlabel(r"$\alpha$")
plt.ylabel(r"$EQM_{boot}(\alpha)$")
plt.legend()
plt.show()
plt.close()
```

## Question 10

```python
def stpforward(X, y, M):
    """Orthogonal Matching Pursuit model (OMP).
    X : Array-like, shape (n_samples, n_features).
    y : Array-like, shape (n_samples, ).
        Target values.
    M : Integer, in [1, n_features]
    """
    selected_variables = []
    residual = y
    p = X.shape[1]
    coef_selected = np.zeros(p)
    for i in range(1, M + 1):
        tab_alphaj = np.zeros(p)
        for j in range(0, p):
            if (j not in selected_variables):
                # Compute Alphaj and add it in tab_Alphaj
                Xj = X[:, j]
                # XXX : Get alpha_j value here
                valeur_alphaj = np.abs(Xj.dot(residual))
                tab_alphaj[j] = valeur_alphaj
        jmax = np.argmax(tab_alphaj)
        selected_variables.append(jmax)
        X_selected = X[:, selected_variables]
        # XXX : perform OLS over selected variables
        skl_linmod = linear_model.LinearRegression()
        skl_linmod.fit(X_selected, y)
        y_pred = skl_linmod.predict(X_selected)
        # Store coefficients
        coef_selected[selected_variables] = skl_linmod.coef_
        # XXX : Update residual
        residual = y - y_pred
    return coef_selected, selected_variables
```

## Question 11

```python
class MYOMP(LinearModel, RegressorMixin):
    """Orthogonal Matching Pursuit model (OMP).
    Parameters
    ----------
    n_nonzero_coefs : int, optional
    Desired number of non-zero entries in the solution. If None (by 
    default) this value is set to 10% of n_features.
    """
    def __init__(self, n_nonzero_coefs=None, fit_intercept=False,
    normalize=False, precompute='auto'):
        self.fit_intercept = False
        self.normalize = normalize
        self.precompute = precompute
        self.n_nonzero_coefs = n_nonzero_coefs
        
    def fit(self, X, y):
        """Fit the model using X, y as training data.
        Parameters
        ----------
        X : array-like, shape (n_samples, n_features)
        Training data.
        y : array-like, shape (n_samples,) or (n_samples, n_targets)
        Target values.
        Returns
        -------
        self : object
        returns an instance of s
        """
        #self.coef_ = np.zeros([X.shape[[1], self.n_nonzero_coefs]])
        # XXX : MODIFY HERE !!!
        coef_selected, selected_variables = stpforward(X, y, self.n_nonzero_coefs)
        
        self.coef_ = coef_selected
        self.selected_variables = selected_variables
        self.intercept_ = 0.
        return self
```
