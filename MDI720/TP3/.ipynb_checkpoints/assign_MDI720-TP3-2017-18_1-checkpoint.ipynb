## Exercice 1. Régression robuste

```python
# Import des librairies

from os import path
import pandas as pd
import sys
import numpy as np
from scipy import stats
from functools import partial

import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

On définit la function $\rho_\alpha(x)$:

```python
def get_rho(X, alpha):
    if (alpha == 0):
        X = abs(X)
    else:
        X = np.where(abs(X) <= alpha, np.power(X, 2) / 2,
                     alpha * abs(X) - np.power(alpha, 2) / 2)
    return X
```

### 1) Génération de X et Y

```python
np.random.seed(1)

# Génération de X uniforme
X = np.random.rand(100, 2)
X_len = X.shape[0]
print("X size: " + str(X.shape))

# Génération de epsilon gaussien centré réduit
epsilon_g = np.random.randn(100, 1)
print("eps size: " + str(epsilon_g.shape))

# Définition de Theta
Theta = np.ones((3, 1))

# Calcul de Y
Y = Theta[0] + np.dot(X, Theta[1:]) + epsilon_g
print("Y size: " + str(Y.shape))
```

### 1) Visualisation de $\rho_\alpha(x)$ pour $\alpha\in\{0.5, 2, 5\}$

```python
# Définition de l'abscisse
x_grid = np.linspace(-5, 5, 1000)
df = pd.DataFrame(x_grid, columns=["x"])

df["y alpha=0.5"] = get_rho(df['x'], 0.5)
df["y alpha=2"] = get_rho(df['x'], 2)
df["y alpha=5"] = get_rho(df['x'], 5)
df.set_index(df.x, inplace=True)
del df["x"]
```

```python
df.plot(figsize=[12, 6])
plt.title(r"$\rho_\alpha(x)$")
legend = plt.legend()
frame = legend.get_frame()
frame.set_color('white')
```

Plus $\alpha$ est grand et plus l'estimateur de $\hat{\theta}$ se rapproche d'un estimateur des moindres carrés. Pour un $\alpha$ petit, on est plus dans le cas d'un LAD régularisé.

### 3) Calculer $\hat{\theta}_\alpha$

```python
# Fonction de minimization
def to_minimize(Theta, alpha, X, Y):
    result = np.sum(
        get_rho(Y - Theta[0] - np.dot(X, Theta[1:].reshape(2, 1)), alpha))
    return result
```

```python
# Minimisation de Theta pour alpha=2
from scipy import optimize
alpha = 2
Theta_init = np.zeros((3, 1))
Opti_result = optimize.minimize(to_minimize, Theta_init, args=(
    alpha, X, Y), method='nelder-mead', options={'xtol': 1e-5})
print("Theta_hat: " + str(Opti_result.x))
```

### 4) Algorithme de boostrap des résidus

```python
# Calcul des résidus
Theta_hat = Opti_result.x.reshape(3, 1)
Residu = Y - Theta_hat[0] - np.dot(X, Theta_hat[1:].reshape((2, 1)))
df_residu = pd.DataFrame(Residu, columns=['epsilon_hat'])

# Visualisation des résidus
df_residu.plot(figsize=[12, 6])
plt.title(r"$\hat{\epsilon_\alpha}$")
legend = plt.legend()
frame = legend.get_frame()
frame.set_color('white')
```

Les résidus obtenus semblent bien centrés avec une variance près de 1.

On définit maintenant la fonction de bootstrap qu'on utilisera par la suite.

```python
# bootstrap
def get_theta_star(df_residu, Theta_hat, X, alpha, B, Theta_out=None):
    Theta_init = np.zeros((3, 1))
    if Theta_out is None:
        Theta_out = []
    for i in range(0, B):
        residu_boot = df_residu.sample(X.shape[0], replace=True).values
        Y_star = Theta_hat[0] + np.dot(X, Theta_hat[1:]) + residu_boot
        Theta_star = optimize.minimize(to_minimize, Theta_init, args=(
            alpha, X, Y_star), method='nelder-mead', options={'xtol': 1e-5})
        Theta_out.append(Theta_star.x)
    return Theta_out
```

Calcul d'un estimateur $\hat{\theta}_\alpha^*$ pour B = 1 et $\alpha$ = 2:

```python
# Initialisation des paramètres
B = 1
alpha = 2

# Lancement du bootstrap
Theta_boot = pd.DataFrame(get_theta_star(
    df_residu, Theta_hat, X, alpha, B), columns=["Theta0", "Theta1", "Theta2"])
Theta_boot
```

### 5) Calcul des statistiques pour 200 réplications de boostrap 

```python
# Initialisation des paramètres
B = 200
alpha = 2

# Lancement du bootstrap
Theta_boot = pd.DataFrame(get_theta_star(
    df_residu, Theta_hat, X, alpha, B), columns=["Theta0", "Theta1", "Theta2"])
Theta_boot.head()
```

Calcul de la matrice $V_{boot}$:

```python
Var_boot = np.cov(Theta_boot.T)
print(Var_boot)
```

Calcul de $b_{boot}$:

```python
Biais_boot = 1 / B * np.sum(Theta_boot - Theta_hat.reshape(3,), axis=0)
print(Biais_boot)
```

Calcul de $EQM_{boot}(\alpha)$:

```python
EQM_boot = 1 / B * \
    np.power(np.linalg.norm(Theta_boot - Theta_hat.reshape(3,)), 2)
EQM_boot
```

### 6) Chercher $\alpha$ qui minimise $EQM_{boot}(\alpha)$

```python
# Définition de la liste des alphas
alphas = np.array(range(0, 21, 1)) / 2
print(alphas)
```

```python
# Définition de la fonction de calcul de EQM_boot pour 1 alpha
def get_EQM_boot(df_residu, Theta_hat, X, alpha):
    B = 200
    Theta_boot = pd.DataFrame(get_theta_star(
        df_residu, Theta_hat, X, alpha, B), columns=["Theta0", "Theta1", "Theta2"])
    EQM_boot = 1 / B * \
        np.power(np.linalg.norm(Theta_boot - Theta_hat.reshape(3,)), 2)
    return EQM_boot
```

```python
# Lancement des boostrap pour chaque alpha
EQM_list_g = [get_EQM_boot(df_residu, Theta_hat, X, alpha) for alpha in alphas]
```

```python
# Visualisation de la courbe
plt.figure(figsize=(12, 6))
plt.plot(alphas, EQM_list_g, label=r"$EQM_{boot}(\alpha)$")
plt.title(r"$EQM_{boot}(\alpha)$")
legend = plt.legend()
frame = legend.get_frame()
frame.set_color('white')

# Sélection du meilleur theta
dict = {}
dict['alphas'] = alphas
dict['EQM_list'] = EQM_list_g
df_g = pd.DataFrame.from_dict(dict)
df_g.sort_values("EQM_list")[:1]
```

### 8) On applique maintenant la méthode précédente au jeu de données "diabetes"

```python
# Import des données
import sklearn.datasets
data = sklearn.datasets.load_diabetes()
```

```python
# On ne récupère que la quatrième valeur expliquative
X = data['data'][:, 3]
X_len = X.shape[0]
X = X.reshape(X_len, 1)
print("X size: " + str(X.shape))

# Définition de Theta
Theta = np.ones((2, 1))

# Définition de Y
Y = data['target'].reshape(X_len, 1)
print("Y size: " + str(Y.shape))
```

On redéfinit la fonction **to_minimize** car $\theta$ est maintenant un scalaire.

```python
# Fonction de minimization
def to_minimize(Theta, alpha, X, Y):
    result = np.sum(
        get_rho(Y - Theta[0] - Theta[1] * X, alpha))
    return result
```

```python
Theta_init = np.zeros((2, 1))
alpha = 2
# Caclul des résidus
Opti_result = optimize.minimize(to_minimize, Theta_init, args=(
    alpha, X, Y), method='nelder-mead', options={'xtol': 1e-5})

Theta_hat = Opti_result.x.reshape(2, 1)
print("Theta_hat: " + str(Opti_result.x))
Residu = Y - Theta_hat[0] - Theta[1] * X
df_residu = pd.DataFrame(Residu, columns=['epsilon_hat'])

# Visualisation des résidus
df_residu.plot(figsize=[12, 6])
plt.title(r"$\hat{\epsilon_\alpha}$")
legend = plt.legend()
frame = legend.get_frame()
frame.set_color('white')
```

Le bruit a l'air gaussien, on s'attend à avoir un $EQM_{boot}$ qui converge vers un minimum pour $\alpha$ assez grand.

De la même façon on redéfinit la fonction **get_theta_star**.

```python
def get_theta_star(df_residu, Theta_hat, X, alpha, B, Theta_out=None):
    Theta_init = np.zeros((2, 1))
    if Theta_out is None:
        Theta_out = []
    for i in range(0, B):
        residu_boot = df_residu.sample(X.shape[0], replace=True).values
        Y_star = Theta_hat[0] + np.dot(X, Theta_hat[1:]) + residu_boot
        Theta_star = optimize.minimize(to_minimize, Theta_init, args=(
            alpha, X, Y_star), method='nelder-mead', options={'xtol': 1e-5})
        Theta_out.append(Theta_star.x)
    return Theta_out
```

```python
# Définition de la fonction de calcul de EQM_boot pour 1 alpha
def get_EQM_boot(df_residu, Theta_hat, X, alpha):
    B = 200
    Theta_boot = pd.DataFrame(get_theta_star(
        df_residu, Theta_hat, X, alpha, B), columns=["Theta0", "Theta1"])
    EQM_boot = 1 / B * \
        np.power(np.linalg.norm(Theta_boot - Theta_hat.reshape(2,)), 2)
    return EQM_boot
```

Nous allons choisir des $\alpha$ compris entre 0 et 5000 sur une échelle logarithmique.

```python
alphas = np.array((np.logspace(0, 4, 20) - 1) / 2)
alphas
```

```python
# Lancement des bootstrap pour chaque alpha
EQM_list_cd = [get_EQM_boot(df_residu, Theta_hat, X, alpha)
               for alpha in alphas]
```

```python
# Visualisation de la courbe
plt.figure(figsize=(12, 6))
plt.plot(alphas, EQM_list_cd, label=r"$EQM_{boot}(\alpha)$")
plt.title(r"$EQM_{boot}(\alpha)$")
legend = plt.legend()
frame = legend.get_frame()
frame.set_color('white')

# Sélection du meilleur theta
dict = {}
dict['alphas'] = alphas
dict['EQM_list'] = EQM_list_cd
df_g = pd.DataFrame.from_dict(dict)
df_g.sort_values("EQM_list")[:1]
```

## Exercice 2. Algorithmes gloutons ou *greedy*
Nous allons dans cet exercice travailler sur la base de données **airquality**.

```python
# Import des données
import statsmodels.datasets as sd
data = sd.get_rdataset('airquality').data
data.head()
```

### 9) Standardiser les données

```python
# On supprime les valeurs manquantes
nb_lignes = data.shape[0]

data = data.dropna(axis=0, how='any')
print(str(nb_lignes - data.shape[0]) +
      " valeurs manquantes ont été supprimées.")

# Standardisation des variables explicatives
from sklearn import preprocessing
scaler = preprocessing.StandardScaler()
features = ["Solar.R", "Wind", "Temp", "Month", "Day"]
X = scaler.fit_transform(data[features])
df_x = pd.DataFrame(X, columns=features)
df_y = data["Ozone"]
```

### 10) Ecrire la fonction de forward selection

```python
import numpy as np
from sklearn.linear_model.base import LinearModel
from sklearn.base import RegressorMixin
import sklearn.linear_model as lm


def stpforward(X, y, M, fit_intercept=False):
    """Orthogonal Matching Pursuit model (OMP).
    X: Array-like, shape (n_samples, n_features).
    Training data.
    y: Array-like, shape (n_samples, ).
    Target values.
    M: Integer, in [1,n_features]
    """
    selected_variables = []
    residual = y
    p = X.shape[1]
    coef_selected = np.zeros(p)
    for i in range(1, M + 1):
        tab_alphaj = np.zeros(p)
        for j in range(0, p):
            if(j not in selected_variables):
                # Compute Alphaj and add it in tab_Alphaj
                Xj = X[:, j]
                # XXX: Get alpha_j value here
                valeur_alphaj = np.abs(Xj.dot(residual))
                tab_alphaj[j] = valeur_alphaj
        jmax = np.argmax(tab_alphaj)
        selected_variables.append(jmax)
        X_selected = X[:, selected_variables]
        # XXX: perform OLS over selected variables
        skl_linmod = lm.LinearRegression(fit_intercept=fit_intercept)
        skl_linmod.fit(X_selected, y)
        # Store coefficients
        coef_selected[selected_variables] = skl_linmod.coef_
        # XXX: Update residual
        r = y - np.dot(X_selected, coef_selected[selected_variables])
    return coef_selected, selected_variables, skl_linmod.intercept_
```

On fait un petit test pour M = 2.

```python
M = 2
test = stpforward(df_x.values, df_y.values, M)
print("Selected_columns:")
for i in range(M):
    print("- " + df_x.columns[test[1]][i] + ": " + str(test[0][test[1][i]]))
```

### 11) Créer une classe MYOMP

```python
class MYOMP(LinearModel, RegressorMixin):
    """Orthogonal Matching Pursuit model (OMP).
    Parameters
    ----------
    n_nonzero_coefs : int, optional
    Desired number of non-zero entries in the solution. If None (by
    default) this value is set to 10% of n_features.
    """

    def __init__(self, n_nonzero_coefs=None, fit_intercept=False,
                 normalize=False, precompute='auto'):
        self.fit_intercept = fit_intercept
        self.normalize = normalize
        self.precompute = precompute
        self.n_nonzero_coefs = n_nonzero_coefs
#         self.n_nonzero_coefs = None

    def fit(self, X, y):
        """Fit the model using X, y as training data.
        Parameters
        ----------
        X : array-like, shape (n_samples, n_features)
        Training data.
        y : array-like, shape (n_samples,) or (n_samples, n_targets)
        Target values.
        Returns
        -------
        self : object
        returns an instance of self.
        """
        self.coef_ = np.zeros([X.shape[1], ])
        self.intercept_ = 0.
        # XXX: MODIFY HERE !!!
        if (self.n_nonzero_coefs is None):
            self.n_nonzero_coefs = X.shape[1] // 10 + 1
        self.coef_, self.selected_variables_, self.intercept_ = stpforward(
            X, y, self.n_nonzero_coefs, fit_intercept=self.fit_intercept)
        return self
```

### 12) Appliquer MYOMP sur M = 1, 2, 3, 4, 5

```python
for i in range(1, 6):
    omp = MYOMP(n_nonzero_coefs=i, fit_intercept=True).fit(
        df_x.values, df_y.values)
    print("M = " + str(i) + " coefs: " + str(omp.coef_))
    print()
```

### 13) Comparer avec OrthogonalMatchingPursuit de sklearn

```python
from sklearn.linear_model import OrthogonalMatchingPursuit

for i in range(1, 6):
    # modélisation de MYOMP
    myOmp = MYOMP(n_nonzero_coefs=i, fit_intercept=True).fit(
        df_x.values, df_y.values)

    # modélisation de OrthogonalMatchingPursuit
    omp = OrthogonalMatchingPursuit(
        n_nonzero_coefs=i).fit(df_x.values, df_y.values)

    print("M = " + str(i) + " coefs: " + str(omp.coef_))
    print("MYOMP et OMP identiques: " +
          str(np.allclose(omp.coef_, myOmp.coef_)))
    print()
```

### 14) Construire un CV (3 folds) pour le choix des variables

Nous allons faire un CV sur le modèle MYOMP avec l'**Intercept** pour pouvoir le comparer avec le modèle OMP. A noter que nous appliquons un **shuffle** avant de créer les folds pour casser tout ordre dans les données.

```python
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import KFold

# définition de la grille de paramètres
grid = {}
grid["n_nonzero_coefs"] = [1, 2, 3, 4, 5]

# définition du Kfold
cv = KFold(n_splits=3, shuffle=True, random_state=111)

# modélisation de MYOMP
myOmp = MYOMP(fit_intercept=True)
myGs = GridSearchCV(myOmp, param_grid=grid, scoring="neg_mean_squared_error",cv=cv)
myResult = myGs.fit(df_x.values, df_y.values)
print("* Méthode d'évaluation:\n{}\n".format(myGs.scorer_))
```

```python
print("Mean test score: " + str(myResult.cv_results_['mean_test_score']))
print("On obtient le meilleur score pour M = " +
      str(grid["n_nonzero_coefs"][myResult.cv_results_['mean_test_score'].argmax()]))
```

### 15) Afficher l'erreur quadratique en fonction de M 

Dans cet exercice nous allons tracer les courbes d'erreur quadratique en fonction de M pour les deux modèles.

```python
fig = plt.figure(figsize=[15, 8])

# MYOMP
mean_test_score = myResult.cv_results_['mean_test_score']
best_M = mean_test_score.argmax()
sigma = myResult.cv_results_["std_test_score"][best_M]

plt.subplot(2, 1, 1)
plt.xticks(grid["n_nonzero_coefs"])
plt.errorbar(grid["n_nonzero_coefs"], mean_test_score, yerr=myResult.cv_results_['std_test_score'],
             fmt='-o', markersize=10, ecolor="g", capsize=10, capthick=1, label="MYOMP")

plt.title("MYOMP")
plt.scatter(best_M + 1, mean_test_score.max(),
            lw=1, facecolors="none", edgecolors='dimgrey', s=1000,
            label="Valeur optimale: " + str(round(mean_test_score.max(), 2)) + r"  $\pm$" + str(round(sigma,2)))

plt.ylabel("neg_mean_squared_error")
legend = plt.legend()
legend.legendHandles[0]._sizes = [150]
frame = legend.get_frame()
frame.set_color('white')


# modélisation de OrthogonalMatchingPursuit
omp = OrthogonalMatchingPursuit()
gs = GridSearchCV(omp, param_grid=grid, scoring="neg_mean_squared_error",cv=cv)
result = gs.fit(df_x.values, df_y.values)

mean_test_score = result.cv_results_['mean_test_score']
best_M = mean_test_score.argmax()
sigma = result.cv_results_["std_test_score"][best_M]

plt.subplot(2, 1, 2)
plt.xticks(grid["n_nonzero_coefs"])
plt.errorbar(grid["n_nonzero_coefs"], result.cv_results_[
             'mean_test_score'], yerr=result.cv_results_['std_test_score'],
             fmt='-o', markersize=10, ecolor="g", capsize=10, capthick=1, label="OrthogonalMatchingPursuit")

plt.title("OrthogonalMatchingPursuit")
plt.scatter(best_M + 1, mean_test_score.max(),
            lw=1, facecolors="none", edgecolors='dimgrey', s=1000,
            label="Valeur optimale: " + str(round(mean_test_score.max(), 2)) + r"  $\pm$" + str(round(sigma,2)))

plt.ylabel("neg_mean_squared_error")
legend = plt.legend()
legend.legendHandles[0]._sizes = [150]
frame = legend.get_frame()
frame.set_color('white')
```
