## Q2

### Cas où on traite l'intégralité des données dans le training

```python
############################################
# ARBRES
############################################

# Q2. Créer un objet 'arbre de décision' en spécifiant le critère de
# classification comme l'indice de gini ou l'entropie, avec la
# fonction 'DecisionTreeClassifier' du module 'tree'.

#dt = tree.DecisionTreeClassifier(criterion='gini')

# Effectuer la classification d'un jeu de données simulées

data = data4
X = data[:, :-1]
Y = data[:, -1]

#dt.fit(X, Y)

# Afficher les scores en fonction du paramètre max_depth

dmax = 20
scores = np.zeros((2,dmax))
trees = {}
plt.close(2)
plt.figure(2, figsize=(15, 10))

for j, crit in enumerate(['gini', 'entropy'], 0):
    for i in range(dmax):
        #print(crit,j)
        dt = tree.DecisionTreeClassifier(criterion=crit, max_depth=i+1)
        dt.fit(X, Y)
        
        trees[(j, i)] = dt
        # error = 1 - dt.score(X, Y)
        scores[j, i] = dt.score(X, Y)

print(scores)

plt.close(3)
plt.figure(3)
plt.plot(np.arange(dmax) + 1, scores[0,], label="Gini") 
plt.plot(np.arange(dmax) + 1, scores[1,], label="Entropy")
plt.xlabel('Max depth')
plt.ylabel('Accuracy Score')
plt.legend()
plt.show()
```

```python
plt.close(3)
plt.figure(3)
plt.plot(np.arange(dmax) + 1, 1-scores[0,], label="Gini") 
plt.plot(np.arange(dmax) + 1, 1-scores[1,], label="Entropy")
plt.xlabel('Max depth')
plt.ylabel('Error')
plt.legend()
plt.show()
```

On obtient de meilleurs scores avec le critère de Gini. 

### Visualisation Gini / Entropie

```python
# Entropie / Gini

x = np.linspace(0.000001, 0.999999, 1000)
plt.plot(x, -x*np.log10(x)-(1-x)*np.log10(1-x), label='Entropy')
plt.plot(x, 2*x*(1-x), label='Gini')
plt.axvline(0.5, linestyle='--', linewidth=1)

# En 0 : très homogène
# En 1 : très homogène
# En 1/2 : impureté maximale

plt.legend()
plt.show()
```

```python
from sklearn.model_selection import train_test_split

# Split data into train set and test set
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, random_state=42)
```

```python
print(X_test.shape)
```

```python
############################################
# ARBRES
############################################

# Afficher les scores en fonction du paramètre max_depth

dmax = 12
scores_test = np.zeros((2,dmax))
plt.close(2)
plt.figure(2, figsize=(15, 10))

for j, crit in enumerate(['gini', 'entropy'], 0):
    for i in range(dmax):
        #print(crit,j)
        dt = tree.DecisionTreeClassifier(criterion=crit, max_depth=i+1)
        dt.fit(X_train, Y_train)
        # error = 1 - dt.score(X, Y)
        scores_test[j, i] = dt.score(X_test, Y_test)

print(scores_test)

plt.close(3)
plt.figure(3)
plt.plot(np.arange(dmax) + 1, scores_test[0,], label="Gini") 
plt.plot(np.arange(dmax) + 1, scores_test[1,], label="Entropy")
plt.xlabel('Max depth')
plt.ylabel('Accuracy Score')
plt.legend()
plt.show()
```

```python
plt.close(3)
plt.figure(3)
plt.plot(np.arange(dmax) + 1, 1 - scores_test[0,], label="Gini") 
plt.plot(np.arange(dmax) + 1, 1 - scores_test[1,], label="Entropy")
plt.xlabel('Max depth')
plt.ylabel('Error')
plt.legend()
plt.show()
```

## Q7

```python
from sklearn.model_selection import cross_val_score

t = tree.DecisionTreeClassifier()
scores = cross_val_score(t, X, Y, cv=5)

depths = np.arange(1, 16) 

scores = list()
scores_std = list()

for depth in depths:
    t.max_depth = depth
    this_scores = cross_val_score(t, X, Y, n_jobs=1)
    scores.append(np.mean(this_scores))
    scores_std.append(np.std(this_scores))

plt.plot(depths, scores, color="crimson", label="Gini") 
plt.plot(depths, np.array(scores) + np.array(scores_std), color="palevioletred", linewidth=1, label='Standard deviation')
plt.plot(depths, np.array(scores) - np.array(scores_std), color="palevioletred", linewidth=1)
plt.fill_between(depths, scores, np.array(scores) + np.array(scores_std), color="palevioletred", alpha=0.2)
plt.fill_between(depths, scores, np.array(scores) - np.array(scores_std), color="palevioletred", alpha=0.2)
plt.xlabel('Max depth')
plt.ylabel('Score')
plt.legend()
plt.show()
```

```python
print("Profondeur d'arbre pour score maximal : ", np.argmax(scores) + 1)
```

## Q8

```python
from sklearn.model_selection import learning_curve

def plot_learning_curve(estimator, title, X, y, ylim=None, cv=None,
                        n_jobs=1, train_sizes=np.linspace(.1, 1.0, 5)):
    """
    Generate a simple plot of the test and training learning curve.

    Parameters
    ----------
    estimator : object type that implements the "fit" and "predict" methods
        An object of that type which is cloned for each validation.

    title : string
        Title for the chart.

    X : array-like, shape (n_samples, n_features)
        Training vector, where n_samples is the number of samples and
        n_features is the number of features.

    y : array-like, shape (n_samples) or (n_samples, n_features), optional
        Target relative to X for classification or regression;
        None for unsupervised learning.

    ylim : tuple, shape (ymin, ymax), optional
        Defines minimum and maximum yvalues plotted.

    cv : int, cross-validation generator or an iterable, optional
        Determines the cross-validation splitting strategy.
        Possible inputs for cv are:
          - None, to use the default 3-fold cross-validation,
          - integer, to specify the number of folds.
          - An object to be used as a cross-validation generator.
          - An iterable yielding train/test splits.

        For integer/None inputs, if ``y`` is binary or multiclass,
        :class:`StratifiedKFold` used. If the estimator is not a classifier
        or if ``y`` is neither binary nor multiclass, :class:`KFold` is used.

        Refer :ref:`User Guide <cross_validation>` for the various
        cross-validators that can be used here.

    n_jobs : integer, optional
        Number of jobs to run in parallel (default 1).
    """
    plt.figure()
    plt.title(title)
    if ylim is not None:
        plt.ylim(*ylim)
    plt.xlabel("Training examples")
    plt.ylabel("Score")
    train_sizes, train_scores, test_scores = learning_curve(
        estimator, X, y, cv=cv, n_jobs=n_jobs, train_sizes=train_sizes)
    train_scores_mean = np.mean(train_scores, axis=1)
    train_scores_std = np.std(train_scores, axis=1)
    test_scores_mean = np.mean(test_scores, axis=1)
    test_scores_std = np.std(test_scores, axis=1)
    plt.grid()

    plt.fill_between(train_sizes, train_scores_mean - train_scores_std,
                     train_scores_mean + train_scores_std, alpha=0.1,
                     color="r")
    plt.fill_between(train_sizes, test_scores_mean - test_scores_std,
                     test_scores_mean + test_scores_std, alpha=0.1, color="g")
    plt.plot(train_sizes, train_scores_mean, 'o-', color="r",
             label="Training score")
    plt.plot(train_sizes, test_scores_mean, 'o-', color="g",
             label="Cross-validation score")

    plt.legend(loc="best")
    return plt
```

```python
title = "Learning Curves (Decision Tree)"

estimator_0 = tree.DecisionTreeClassifier()
plot_learning_curve(estimator_0, title, X, Y, ylim=(0, 1.01), cv=5, n_jobs=-1)
```

```python
title = "Learning Curves (Decision Tree)"

estimator = tree.DecisionTreeClassifier(max_depth=9)
plot_learning_curve(estimator, title, X, Y, ylim=(0, 1.01), cv=5, n_jobs=-1)
```

If the training score is much greater than the validation score for the maximum number of training samples, adding more training samples will most likely increase generalization. In the above plot you can see that the DecisionTree could benefit from more training examples.

## Q9

```python
from sklearn import linear_model

data = data4
X = data[:, :-1]
Y = data[:, -1]

# Split data into train set and test set
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, random_state=42)

lr = linear_model.LogisticRegression(fit_intercept=True)
lr.fit(X_train, Y_train)
scores_lr = lr.score(X_test, Y_test)

print("Score : ", scores_lr)
print("Error : ", 1 - scores_lr)
print()
print('coef_ :  (n_classes, n_features)')
print(lr.coef_)

print()
print('intercept_ : (n_classes, )')
print(lr.intercept_)
```

```python
frontiere_new(lambda x: lr.predict(x.reshape((1, -1))), X_test, Y_test, step=100)
plt.title('Logistic Regression (fit_intercept = True)')
plt.show()
```

Les résultats sont bien moins bons que ceux obtenus avec l'arbre de décision. Cela peut s'expliquer par le fait que le modèle considéré n'est pas linéairement séparable : par exemple les points oranges sont dispersés non regroupés. Ils sont de part et d'autre des autres classes. Une seule droite ne peut donc pas classifier de manière satisfaisante.

```python
lr.fit_intercept = False 
lr.fit(X_train, Y_train)
frontiere_new(lambda x: lr.predict(x.reshape((1, -1))), X_test, Y_test, step=100)
plt.title('Logistic Regression (fit_intercept = False)')
plt.show()

print()
print('coef_ :  (n_classes, n_features)')
print(lr.coef_)

print()
print('intercept_ : (n_classes, )')
print(lr.intercept_)
```
